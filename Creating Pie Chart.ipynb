```python
import matplotlib.pyplot as plt
```

```python
fig=plt.figure(figsize=(8,7))
perc=[50,20,15,15]
dept=["Statistics","Maths","Computers","Physics"]
cols=['magenta','cyan','brown','gold']
plt.pie(perc,labels=dept,colors=cols,explode=(0,0.2,0,0),shadow=True,autopct="%.1f%%")
plt.title("JNTU University")
plt.legend()
plt.show()
```
