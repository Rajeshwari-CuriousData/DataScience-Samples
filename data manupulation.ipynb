```python
import pandas as pd
```

```python
df1=pd.read_csv("https://raw.githubusercontent.com/harissriramoju/DataScience-Samples/master/concat_1.csv")
df2=pd.read_csv("https://raw.githubusercontent.com/harissriramoju/DataScience-Samples/master/concat_2.csv")
df3=pd.read_csv("https://raw.githubusercontent.com/harissriramoju/DataScience-Samples/master/concat_3.csv")
```

```python
# concat function from pandas

row_wise=pd.concat([df1,df2,df3],axis=0) # here axis=0 indicating row_wise data
row_wise
```

```python
column_wise=pd.concat([df1,df2,df3],axis=1)  # here axis=1 indicating column wise data                                
column_wise                                             
```

```python
column_wise.loc[3]     # by using loc we are getting row 3 number data
```

```python
row_wise.loc[2]        # by using loc we are getting all the row 2 numbers data
```

```python
row_wise.iloc[3]   # by using iloc we are getting idex number data
```
